Script per a partir d'un txt on es posa el print que es realitza al anterior notebook, es calculen la mitjana, la mediana i la moda entera de les distàncies dels valors positius i negatius per un posterior anàlisis

```python
import re
import statistics

def extract_misclassified_data(file_path):
    distances_true_1 = []  
    distances_true_0 = []
    distances_correcte_1 = []
    distances_correcte_0 = []  
    
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
       # extreure les dades de les línies que contenen la informació de les prediccions
        pattern = re.compile(
            r"Sentence: (.*?)\nTrue Label: (\d+), Predicted Label: (\d+)\nScores: Positive: ([\d.]+), Negative: ([\d.]+)",
            re.DOTALL
        )
        
        matches = pattern.findall(content)
        
        for sentence, true_label, predicted_label, pos_score, neg_score in matches:
            true_label, predicted_label = int(true_label), int(predicted_label)
            pos_score, neg_score = float(pos_score), float(neg_score)
            
            distance = abs(pos_score - neg_score)
            if true_label != predicted_label:
                if true_label == 1:
                    distances_true_1.append(distance)
                elif true_label == 0:
                    distances_true_0.append(distance)
            
            else:
                if true_label == 1:
                    distances_correcte_1.append(distance)
                elif true_label == 0:
                    distances_correcte_0.append(distance)

    
    def compute_statistics(distances):
        if not distances:
            return 0, 0, 0
        
        avg = sum(distances) / len(distances)
        med = statistics.median(distances)
        
        int_distances = [int(d) for d in distances]
        try:
            mode_val = statistics.mode(int_distances)
        except statistics.StatisticsError:
            mode_val = None 
        
        return avg, med, mode_val

    stats_true_1 = compute_statistics(distances_true_1)
    stats_true_0 = compute_statistics(distances_true_0)
    stats_correcte_1 = compute_statistics(distances_correcte_1)
    stats_correcte_0 = compute_statistics(distances_correcte_0)

    return stats_true_1, stats_true_0, stats_correcte_1, stats_correcte_0

file_path = "llistat_errors.txt"  
stats_1, stats_0, stats_cor_1, stats_cor_0 = extract_misclassified_data(file_path)

print('Anàlisi de les classificacions errònies: \n')
print(f"True Label = 1, Error de classificació:")
print(f"  - Distància mitjana: {stats_1[0]}")
print(f"  - Distància mediana: {stats_1[1]}")
print(f"  - Distància moda (int): {stats_1[2]}")

print(f"\nTrue Label = 0, Error de classificació:")
print(f"  - Distància mitjana: {stats_0[0]}")
print(f"  - Distància mediana: {stats_0[1]}")
print(f"  - Distància moda (int): {stats_0[2]}")

print()
print('Anàlisi de les classificacions correctes: \n')
file_path = "llistat_correctes.txt"  
stats_1, stats_0, stats_cor1, stats_cor0 = extract_misclassified_data(file_path)

print(f"True Label = 1, Classificació correcta:")
print(f"  - Distància mitjana: {stats_cor1[0]}")
print(f"  - Distància mediana: {stats_cor1[1]}")
print(f"  - Distància moda (int): {stats_cor1[2]}")

print(f"\nTrue Label = 0, Classificació correcta:")
print(f"  - Distància mitjana: {stats_cor0[0]}")
print(f"  - Distància mediana: {stats_cor0[1]}")
print(f"  - Distància moda (int): {stats_cor0[2]}")

```
